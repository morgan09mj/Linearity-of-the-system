```python
def plot(x, title='Plot of DT signal x[n]'):
  """
  x: DT signal x[n], 1D array
  """
  n = np.arange(x.shape[0])
  plt.xlabel('n');
  plt.ylabel('x[n]');
  plt.title(title);
  plt.plot(n, x, '.-');
  plt.show()
```
